```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.datasets import imdb
```

```python
(train_data, train_targets), (test_data, test_targets) = imdb.load_data(num_words = 10000)
```

```python
def vectorize_sequence(seq, dim = 10000):
  res = np.zeros((len(seq), dim))
  for index, data in enumerate(seq):
    res[index, data] = 1
  return res
```

```python
train_data = vectorize_sequence(train_data)
test_data = vectorize_sequence(test_data)
```

```python
y_train = np.asarray(train_targets).astype('float32')
y_test = np.asarray(test_targets).astype('float32')
```

```python
model = Sequential()
```

```python
model.add(Dense(32, input_shape = (10000, ), activation = 'relu'))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(optimizer = 'adam', loss = 'mse', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(train_data, test_data, epochs = 20, batch_size = 512, validation_split = 0.3)
```

```python
evaluation = model.evaluate(test_data, test_targets)
```
